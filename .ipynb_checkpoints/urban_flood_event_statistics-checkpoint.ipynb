```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# --- user input ---
CSV_PATH = r"D:\Development\RESEARCH\urban_flood_database\event_datatables\urban_flood_database.csv"

# Load data
flood_db = pd.read_csv(CSV_PATH, low_memory=False)

# Ensure numeric types for relevant columns
num_cols = [
    'duration_days',
    'dfo_dead',
    'dfo_displaced',
    'urban_area_km2',
    'urban_flood_area_km2',
    'flood_share_urban_pct'
]
for c in num_cols:
    flood_db[c] = pd.to_numeric(flood_db[c], errors='coerce')

# Parse time for summaries (keep original string format in the CSV)
flood_db['_time_start'] = pd.to_datetime(flood_db['time_start_utc'], errors='coerce')

# Basic counts
n_rows = len(flood_db)                                # total records (clusters)
n_gfd_events = flood_db['gfd_event_id'].nunique()     # unique GFD events
n_cities = flood_db['city_id'].nunique()              # unique cities

print(f"Total rows (clusters): {n_rows:,}")
print(f"Unique GFD events: {n_gfd_events:,}")
print(f"Unique cities: {n_cities:,}")

# Optional: quick temporal range summary
start_min = flood_db['_time_start'].min()
start_max = flood_db['_time_start'].max()
print(f"Date range: {start_min} .. {start_max}")

# Drop helper
flood_db = flood_db.drop(columns=['_time_start'])
flood_db
```

```python


# --- Events per year ---
flood_db['year'] = pd.to_datetime(flood_db['time_start_utc'], errors='coerce').dt.year

# count unique events per year
year_counts = (
    flood_db.drop_duplicates('event_id')['year']
    .value_counts()
    .sort_index()
)

plt.figure(figsize=(10, 5))
plt.bar(year_counts.index, year_counts.values, color='steelblue', edgecolor='black')

# ensure integer ticks only (no 2.5-year spacing)
years = np.arange(year_counts.index.min(), year_counts.index.max() + 1)
plt.xticks(years, rotation=45)

plt.title("Number of flood events per year")
plt.xlabel("Year")
plt.ylabel("Event count")
plt.grid(alpha=0.3, axis='y')
plt.tight_layout()
plt.show()
```

```python
# --- Flooded area ratio distribution ---
plt.figure(figsize=(8,5))
flood_db['flood_share_urban_pct'].plot.hist(bins=30, edgecolor='black')
plt.title("Distribution of flooded urban area share (%)")
plt.xlabel("Flooded area share (%)")
plt.ylabel("Count")
plt.grid(alpha=0.3)
plt.show()
```

```python
# --- Average flooded share by year ---
yearly_mean = flood_db.groupby('year')['flood_share_urban_pct'].mean().dropna()

plt.figure(figsize=(10, 5))
plt.plot(yearly_mean.index, yearly_mean.values, marker='o', color='steelblue', linestyle='-', markerfacecolor='red')

# Ensure integer ticks for years
years = np.arange(yearly_mean.index.min(), yearly_mean.index.max() + 1)
plt.xticks(years, rotation=45)

plt.title("Average flooded share (%) per year")
plt.xlabel("Year")
plt.ylabel("Mean flooded share (%)")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# --- Severity and causes summary ---
severity_counts = flood_db.drop_duplicates('event_id')['dfo_severity'].value_counts()
cause_counts = flood_db.drop_duplicates('event_id')['dfo_main_cause'].value_counts()

print("\nTop flood causes:")
cause_counts
```

```python
print("\nSeverity distribution:")
severity_counts
```

```python
# --- Optional: Summary statistics for numeric fields ---
print("\nSummary of numeric variables:")
flood_db[num_cols].describe().T.round(2)
```

```python
# Boxplots for numeric variables
flood_db[num_cols].plot(kind='box', subplots=True, layout=(2, 4), figsize=(14, 6), sharex=False, sharey=False)
plt.suptitle("Boxplots of numeric variables")
plt.tight_layout()
plt.show()
```
