# Chronicle Flood Database Intersection with MERIT Hydro dataset

This notebook provides a comprehensive analysis of the Chronicle urban flood dataset, containing over 880,000 flood events worldwide from 2000-2025.


```python
import pandas as pd
import numpy as np
from matplotlib.lines import Line2D
import geopandas as gpd
import folium
import branca.colormap as cm
from shapely import wkt
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


ENRICHED_PKL = r"D:\Development\RESEARCH\urban_flood_database\chronicle\hydromerit_pluvial_outputs\chronicle_df_with_pfdi_FULL.pkl"
df_pfdi_clean = pd.read_pickle(ENRICHED_PKL)

print(f"Total rows loaded: {len(df_pfdi_clean)}")

# Quick sanity check
needed_cols = ["upa_p95", "upa_p99", "upa_max", "poly_area_km2", "PFDI_p95", "PFDI_p99", "PFDI_max"]
missing = [c for c in needed_cols if c not in df_pfdi_clean.columns]
if missing:
    raise ValueError(f"Missing required columns in enriched file: {missing}")

print("\nSample rows:")
print(df_pfdi_clean[["upa_max", "upa_p95", "upa_p99", "PFDI_max", "PFDI_p99", "PFDI_p95"]].head())
```

```python
import matplotlib.pyplot as plt
# =========================
# Plotting
# =========================
plt.figure(figsize=(10, 8))

# Define X axis data
x = df_pfdi_clean["poly_area_km2"]

# Plot UPA Max (Red) - Plotted first as it is likely the largest value
# plt.scatter(
#     x, 
#     df_pfdi_clean["upa_max"], 
#     color='red', 
#     alpha=0.5, 
#     label='UPA Max', 
#     s=15
# )

# Plot UPA p99 (Orange)
plt.scatter(
    x, 
    df_pfdi_clean["upa_p99"], 
    color='orange', 
    alpha=0.5, 
    label='UPA p99', 
    s=15
)

# # Plot UPA p95 (Blue) - Plotted last to be on top
# plt.scatter(
#     x, 
#     df_pfdi_clean["upa_p95"], 
#     color='blue', 
#     alpha=0.5, 
#     label='UPA p95', 
#     s=15
# )

# =========================
# Formatting
# =========================
plt.xlabel("Polygon Area (km)")
plt.ylabel("Upstream Drainage Area (UPA) [km]")
plt.title("Polygon Area vs. UPA Statistics (Max, p99, p95)")
plt.legend()
plt.grid(True, which="both", linestyle="--", alpha=0.6)

# Log-log scale is essential for hydrological area data
plt.xscale("log")
plt.yscale("log")

# Optional: Add a 1:1 identity line for reference
# Points below this line have UPA < Area (Local rain / Pluvial)
# Points above this line have UPA > Area (River flow / Fluvial)
limit_min = min(df_pfdi_clean["poly_area_km2"].min(), df_pfdi_clean["upa_p95"].min())
limit_max = max(df_pfdi_clean["poly_area_km2"].max(), df_pfdi_clean["upa_max"].max())
plt.plot([limit_min, limit_max], [limit_min, limit_max], color='black', linestyle='--', linewidth=1, label="1:1 Line")

plt.show()
```

```python
# ==========================================
# 0. CONFIGURATION & SUBSETTING
# ==========================================
AREA_THRESHOLD = 500  # Threshold for polygon area in km^2

# Filter the main dataframe for the specific analysis threshold
df_subset = df_pfdi_clean[df_pfdi_clean["poly_area_km2"] <= AREA_THRESHOLD].copy()

print(f"Plotting analysis for polygons <= {AREA_THRESHOLD} km")
print(f"Number of events in subset: {len(df_subset)}")

# ==========================================
# 1. Plot 1: Histograms
# ==========================================
plt.figure(figsize=(12, 7))

# Define log-spaced bins based on the new PFDI columns
min_val = min(df_subset["PFDI_p95"].min(), df_subset["PFDI_max"].min())
max_val = max(df_subset["PFDI_p95"].max(), df_subset["PFDI_max"].max())

# Safety check for log scale
if min_val <= 0: min_val = 0.001 
bins = np.logspace(np.log10(min_val), np.log10(max_val), 50)

# Plot distributions using the PFDI columns
plt.hist(df_subset["PFDI_max"], bins=bins, alpha=0.4, color='red', label='Index (Max)', edgecolor='black')
plt.hist(df_subset["PFDI_p99"], bins=bins, alpha=0.4, color='orange', label='Index (P99)', edgecolor='black')
plt.hist(df_subset["PFDI_p95"], bins=bins, alpha=0.4, color='blue', label='Index (P95)', edgecolor='black')

# Formatting Plot 1
plt.xscale("log")
plt.xlabel("Index Value (UPA / Area) [Log Scale]")
plt.ylabel("Count of Events")
plt.title(f"Distribution of Flood Indices (Polygons $\leq$ {AREA_THRESHOLD} km)")
plt.legend()
plt.grid(True, which="both", linestyle="--", alpha=0.5)
plt.axvline(x=1, color='black', linestyle='--', linewidth=2, label="Index = 1")
plt.show()

# ==========================================
# 2. Plot 2: Boxplot with Percentages
# ==========================================
def get_pct_below_one(series):
    return (series <= 1).mean() * 100

pct_p95 = get_pct_below_one(df_subset["PFDI_p95"])
pct_p99 = get_pct_below_one(df_subset["PFDI_p99"])
pct_max = get_pct_below_one(df_subset["PFDI_max"])

plt.figure(figsize=(12, 8))

# Data organized for plotting
data_to_plot = [df_subset["PFDI_p95"], df_subset["PFDI_p99"], df_subset["PFDI_max"]]
labels = [
    f'Index (P95)\n({pct_p95:.1f}% $\leq$ 1)', 
    f'Index (P99)\n({pct_p99:.1f}% $\leq$ 1)', 
    f'Index (Max)\n({pct_max:.1f}% $\leq$ 1)'
]

# Define style properties
median_props = dict(linestyle='-', linewidth=2, color='orange')
mean_props = dict(marker='D', markeredgecolor='black', markerfacecolor='green', markersize=8)
box_props = dict(linestyle='-', linewidth=1.5)

bplot = plt.boxplot(data_to_plot, labels=labels, patch_artist=True, 
                    showmeans=True, meanprops=mean_props, 
                    medianprops=median_props, boxprops=box_props)

# Assign colors
colors = ['lightblue', 'bisque', 'mistyrose']
for patch, color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color)

# Formatting Plot 2
plt.yscale("log")
plt.axhline(y=1, color='black', linestyle='--', linewidth=1.5, alpha=0.7)
plt.ylabel("Index Value (UPA / Area) [Log Scale]", fontsize=12)
plt.title(f"Flood Index Distribution & Pluvial Share\n(Polygons $\leq$ {AREA_THRESHOLD} km)", fontsize=14)

legend_elements = [
    Line2D([0], [0], color='orange', lw=2, label='Median'),
    Line2D([0], [0], marker='D', color='w', markerfacecolor='green', markeredgecolor='black', markersize=8, label='Mean'),
    Line2D([0], [0], color='black', linestyle='--', label='1:1 Ratio (Threshold)')
]
plt.legend(handles=legend_elements, loc='upper right')
plt.grid(True, which="both", axis="y", linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()
```

```python

# ============================================
# 1) PREPARE BINS DYNAMICALLY
# ============================================
# Setting bins from 0 up to our defined AREA_THRESHOLD
# The last bin label automatically updates based on your threshold
bins = [0, 10, 50, 100, 250, AREA_THRESHOLD]
labels = ['< 10', '10-50', '50-100', '100-250', f'250-{AREA_THRESHOLD}']

# Create the categorical column based on the threshold
# We use df_subset which was already filtered by AREA_THRESHOLD
df_subset['Area_Bin'] = pd.cut(
    df_subset['poly_area_km2'], 
    bins=bins, 
    labels=labels
)

# ============================================
# 2) PLOTTING - Hexbin & Categorical Boxplot
# ============================================
fig, axes = plt.subplots(2, 1, figsize=(12, 14))

# --- PLOT A: Hexbin Density Plot (Area vs PFDI Max) ---
# Visualizing the concentration of events across area and ratio
hb = axes[0].hexbin(
    df_subset["poly_area_km2"], 
    df_subset["PFDI_max"],  # Updated to use PFDI_max
    gridsize=40, 
    xscale='log', 
    yscale='log', 
    cmap='inferno_r', 
    mincnt=1
)

# Add a colorbar to indicate density
cb = fig.colorbar(hb, ax=axes[0])
cb.set_label('Count of Events')

# Pluvial/Fluvial Threshold line at Ratio = 1
axes[0].axhline(y=1, color='cyan', linestyle='--', linewidth=2, label='PFDI=1 (Threshold)')

axes[0].set_title(f"Hexbin Density: Polygon Area vs. PFDI Max (Up to {AREA_THRESHOLD} km)", fontsize=14)
axes[0].set_xlabel("Polygon Area [km] (Log Scale)", fontsize=12)
axes[0].set_ylabel("PFDI Max (UPA Max / Area) [Log Scale]", fontsize=12)
axes[0].legend()
axes[0].grid(True, which="both", linestyle="--", alpha=0.3)


# --- PLOT B: Boxplots by Area Bin ---
# Analyzing the statistical spread of the ratio within each size class
sns.boxplot(
    data=df_subset, 
    x='Area_Bin', 
    y='PFDI_max',  # Updated to use PFDI_max
    ax=axes[1], 
    palette="coolwarm",
    showfliers=False # Outliers are hidden to focus on the interquartile range (IQR)
)

# Reference line for Ratio = 1
axes[1].axhline(y=1, color='black', linestyle='--', linewidth=2, label='PFDI=1')

# Set Y axis to log scale to accommodate hydrological variance
axes[1].set_yscale("log")

axes[1].set_title(f"PFDI Max Distribution by Area Group (Threshold: {AREA_THRESHOLD} km)", fontsize=14)
axes[1].set_xlabel("Area Range [km]", fontsize=12)
axes[1].set_ylabel("PFDI Max (Log Scale)", fontsize=12)
axes[1].grid(True, axis='y', linestyle="--", alpha=0.5)

plt.tight_layout()
plt.show()
```

```python

# 1) PREPARE GEODATAFRAME FROM EXISTING DF
# ============================================
# Work with the cleaned dataframe from previous steps
# We convert WKT to actual geometry objects for mapping
if "geometry" not in df_pfdi_clean.columns and "geometry_wkt" in df_pfdi_clean.columns:
    df_pfdi_clean["geometry"] = df_pfdi_clean["geometry_wkt"].apply(wkt.loads)

# Create GeoDataFrame
gdf = gpd.GeoDataFrame(df_pfdi_clean, geometry="geometry", crs="EPSG:4326")

# Filter by the same threshold used in the plots
gdf_filtered = gdf[gdf["poly_area_km2"] <= AREA_THRESHOLD].copy()

# Choose which metric to visualize on the map (e.g., PFDI_max or PFDI_p95)
VIS_METRIC = "PFDI_max" 
gdf_filtered["logPFDI"] = np.log10(gdf_filtered[VIS_METRIC])

# ============================================
# 2) STRATIFIED SAMPLING FOR PERFORMANCE
#    Map markers can be slow, so we pick high/mid/low samples per size category
# ============================================
def get_strata_sample(df_in, n=15):
    if len(df_in) < (n * 3): return df_in
    high = df_in.nlargest(n, VIS_METRIC)
    low = df_in.nsmallest(n, VIS_METRIC)
    mid = df_in.iloc[(df_in[VIS_METRIC] - df_in[VIS_METRIC].median()).abs().argsort()[:n]]
    return pd.concat([high, mid, low])

# Define size bins based on polygon area
s_small = gdf_filtered[gdf_filtered["poly_area_km2"] < 10]
s_med   = gdf_filtered[(gdf_filtered["poly_area_km2"] >= 10) & (gdf_filtered["poly_area_km2"] < 100)]
s_large = gdf_filtered[gdf_filtered["poly_area_km2"] >= 100]

# Combine samples
gdf_vis = pd.concat([
    get_strata_sample(s_small),
    get_strata_sample(s_med),
    get_strata_sample(s_large)
]).drop_duplicates(subset="event_id")

print(f"Mapping {len(gdf_vis)} representative events using {VIS_METRIC}...")

# ============================================
# 3) COLORMAP LOGIC
#    Log10(1) = 0 is the white point (Equal area vs drainage)
# ============================================
vmin, vmax = -2, 2 # Range for log10(PFDI)
cmap = cm.LinearColormap(
    colors=['red', 'orange', 'white', 'lightblue', 'blue'],
    index=[vmin, -0.5, 0, 0.5, vmax],
    vmin=vmin, vmax=vmax
)
cmap.caption = f"Index: {VIS_METRIC} | Blue: Fluvial (Ratio > 1) | Red: Pluvial (Ratio < 1)"

# ============================================
# 4) BUILD THE MAP
# ============================================
# Center map on the data
center = gdf_vis.geometry.centroid.unary_union.centroid
m = folium.Map(location=[center.y, center.x], zoom_start=4, tiles="cartodbpositron")

def style_fn(feature):
    val = feature["properties"]["logPFDI"]
    val = max(min(val, vmax), vmin) # Clamp
    return {
        "fillColor": cmap(val),
        "color": "black",
        "weight": 0.5,
        "fillOpacity": 0.6
    }

# Add Layer
folium.GeoJson(
    gdf_vis,
    style_function=style_fn,
    tooltip=folium.GeoJsonTooltip(
        fields=["event_id", VIS_METRIC, "poly_area_km2", "upa_max"],
        aliases=["ID:", "PFDI Ratio:", "Area (km):", "Max UPA (km):"],
        localize=True
    )
).add_to(m)

m.add_child(cmap)

# Display Map
m
```
