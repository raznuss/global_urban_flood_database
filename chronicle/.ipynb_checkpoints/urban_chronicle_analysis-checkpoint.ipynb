# Urbanizing the Data: Defining Urban Flood Events

In this section, we filter the dataset to focus on urban environments. 
We set a threshold of **> 20%** built-up area to define an event as "urban".

This threshold was selected to align with the **"developed, low-intensity" class** (defined as **2049%** impervious surfaces) in the **US National Land Cover Database (NLCD)**.

```python
# Filter with parentheses to handle operator precedence correctly
threshold = 20 # parentheses
thresh_chronicle_urban_df = chronicle_urban_df[
    (chronicle_urban_df['urban_percentage'] > threshold) & 
    (chronicle_urban_df['urban_percentage'] <= 100)
]
thresh_chronicle_urban_df
```

```python
# Extract coordinates from geometry for geographic analysis
def extract_centroid_coordinates(geometry_wkt_string):
    """Extract centroid coordinates from WKT geometry string."""
    try:
        if geometry_wkt_string.startswith(('POLYGON', 'MULTIPOLYGON')):
            geom = wkt.loads(geometry_wkt_string)
            centroid = geom.centroid
            return centroid.y, centroid.x  # lat, lon
    except Exception:
        return None, None
    return None, None

# Process sample for geographic visualization
sample_size = len(thresh_chronicle_urban_df)
sample_events = thresh_chronicle_urban_df.sample(n=sample_size, random_state=42)

print(f"Processing {sample_size} events for geographic analysis...")

coordinates = []
areas = []
durations = []

for idx, row in sample_events.iterrows():
    lat, lon = extract_centroid_coordinates(row['geometry_wkt'])
    if lat is not None and lon is not None:
        coordinates.append([lon, lat])  # lon, lat order
        areas.append(row['area_km2'])
        durations.append(row['duration_days'])

coordinates = np.array(coordinates)
lons = coordinates[:, 0]
lats = coordinates[:, 1]
areas_array = np.array(areas)

print(f"Successfully processed {len(coordinates)} events")
```

###  Helper Functions 

```python
# --- Helper Functions ---
def get_pixels_from_projected_geom(geom_projected, raster_path):
    """Extracts pixels from the raster that intersect with the geometry."""
    pixel_geoms, pixel_percentages = [], []

    with rasterio.open(raster_path) as src:
        # Define window based on geometry bounds
        minx, miny, maxx, maxy = geom_projected.bounds
        window = rasterio.windows.from_bounds(minx, miny, maxx, maxy, transform=src.transform)
        
        # Read data (masked)
        data = src.read(1, window=window, boundless=True, fill_value=src.nodata)
        window_transform = src.window_transform(window)
        height, width = data.shape
        
        for row in range(height):
            for col in range(width):
                val = data[row, col]
                if val >= 0 and val != src.nodata:
                    # Create pixel polygon
                    x1, y1 = window_transform * (col, row)
                    x2, y2 = window_transform * (col + 1, row + 1)
                    pixel_poly = box(min(x1, x2), min(y1, y2), max(x1, x2), max(y1, y2))
                    
                    if pixel_poly.intersects(geom_projected):
                        # Calculate pct (assuming val is area coverage or density)
                        # Adjust calculation logic if 'val' represents something else in your specific raster
                        pct = val # Assuming GHS value is already meaningful, or calculate relative to pixel area
                        pixel_geoms.append(pixel_poly)
                        pixel_percentages.append(pct)

    if not pixel_percentages: 
        return gpd.GeoDataFrame()
    
    return gpd.GeoDataFrame(
        {'urban_pct': pixel_percentages}, 
        geometry=pixel_geoms, 
        crs=src.crs
    )

def get_pct_color(value):
    """Colormap: Green (0) -> Red (100)"""
    norm_val = min(max(value, 0), 100) / 100.0
    cmap = plt.get_cmap('RdYlGn_r') 
    return mcolors.to_hex(cmap(norm_val))
```

###  Lets plot


```python
# 1. Select INDEX
SELECTION_INDEX = 0  

if len(thresh_chronicle_urban_df) > SELECTION_INDEX:
    # Get the row using iloc (positional indexing)
    row = thresh_chronicle_urban_df.iloc[SELECTION_INDEX]
    
    print(f"Showing event #{SELECTION_INDEX} from filtered list")
    print(f"Original DB Index: {row.name}")
    print(f"Urban Percentage: {row['urban_percentage']:.2f}%")

    # 2. Geometry & Projection Handling
    # We need to project the WKT (WGS84) to the Raster's CRS to get pixels, then back to WGS84 for the Map.
    
    # Load WKT to Geometry
    geom_wgs84 = wkt.loads(row['geometry_wkt'])
    gs_wgs84 = gpd.GeoSeries([geom_wgs84], crs="EPSG:4326")
    
    # Get Raster CRS and Project
    with rasterio.open(ghs_urban_raster_path) as src:
        raster_crs = src.crs
    
    gs_projected = gs_wgs84.to_crs(raster_crs)
    geom_projected = gs_projected.iloc[0]

    # 3. Extract Pixels
    print("Extracting urban pixels...")
    pixel_gdf_projected = get_pixels_from_projected_geom(geom_projected, ghs_urban_raster_path)

    # 4. Prepare Map Layers (Convert pixels back to WGS84)
    pixel_gdf_wgs84 = None
    if not pixel_gdf_projected.empty:
        pixel_gdf_wgs84 = pixel_gdf_projected.to_crs(epsg=4326)

    # 5. Create Map
    centroid = geom_wgs84.centroid
    m = folium.Map(location=[centroid.y, centroid.x], zoom_start=13, tiles="OpenStreetMap")
    folium.TileLayer('https://server.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{z}/{y}/{x}', 
                     attr='Esri', name='Satellite').add_to(m)

    # Add Flood Boundary (Blue Outline)
    folium.GeoJson(
        gs_wgs84,
        name="Flood Boundary",
        style_function=lambda x: {'color': 'blue', 'weight': 3, 'fillOpacity': 0}
    ).add_to(m)

    # Add Urban Pixels (Colored Squares)
    if pixel_gdf_wgs84 is not None:
        folium.GeoJson(
            pixel_gdf_wgs84,
            name="Urban Raster Data",
            style_function=lambda feature: {
                'fillColor': get_pct_color(feature['properties']['urban_pct']),
                'color': 'none',        
                'fillOpacity': 0.6, 
            },
            tooltip=folium.GeoJsonTooltip(fields=['urban_pct'], aliases=['Urban %:'], localize=True)
        ).add_to(m)

    folium.LayerControl().add_to(m)
    display(m)

else:
    print(f"Error: Index {SELECTION_INDEX} is out of range. The filtered DF has {len(thresh_chronicle_urban_df)} rows.")
```
