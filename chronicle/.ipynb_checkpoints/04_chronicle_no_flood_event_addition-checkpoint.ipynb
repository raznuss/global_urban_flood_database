```python
# =================================================================
# 04_chronicle_no_flood_stratified_sampling
# Goal: Create 18 refined buckets (3x3x2) for targeted sampling.
# Action: Pick representative polygons for GEE No-Flood search.
# =================================================================

import pandas as pd
import numpy as np
import os

# --- 1. CONFIGURATION ---
INPUT_MASTER_PATH = r"D:\Development\RESEARCH\urban_flood_database\chronicle\imerg_rain_outputs\chronicle_rain_master.pkl"
OUTPUT_DIR = r"D:\Development\RESEARCH\urban_flood_database\chronicle\no_flood_event_outputs"
TARGETS_PATH = os.path.join(OUTPUT_DIR, "sampling_targets_for_gee.pkl")

if not os.path.exists(OUTPUT_DIR):
    os.makedirs(OUTPUT_DIR)

# --- 2. LOAD DATA ---
print(f"Loading master dataset: {INPUT_MASTER_PATH}")
master_df = pd.read_pickle(INPUT_MASTER_PATH)

# Focusing on modern era (2018+) for consistency with IMERG and Urban data
modern_events = master_df[master_df['start_time'] >= '2018-01-01'].copy()
print(f"Events available for sampling: {len(modern_events)}")

# --- 3. REFINED STRATIFICATION (18 BUCKETS) ---
print("Categorizing into 18 strategic buckets...")

# A. Area (Based on Distribution - Tertiles)
modern_events['area_bin'] = pd.qcut(modern_events['area_km2'], 3, labels=['Small', 'Medium', 'Large'])

# B. Urban Percentage (Manual Thresholds)
# Thresholds: 0-20% (Low), 20-60% (Urban), 60-100% (Highly Urban)
urban_bins = [-0.01, 20, 60, 100.01]
urban_labels = ['Low-Urban', 'Urban', 'Highly-Urban']
modern_events['urban_bin'] = pd.cut(modern_events['urban_percentage'], bins=urban_bins, labels=urban_labels)

# C. Flood Mechanism (PFDI Threshold - Pluvial vs Fluvial Candidate)
# Threshold: < 1 (Pluvial-leaning), >= 1 (Fluvial/Extreme)
pfdi_bins = [-0.01, 1, np.inf]
pfdi_labels = ['Pluvial (PFDI<1)', 'Fluvial (PFDI>=1)']
modern_events['mechanism_bin'] = pd.cut(modern_events['PFDI_p99'], bins=pfdi_bins, labels=pfdi_labels)

# --- 4. STRATIFIED SAMPLING ---
# Goal: Sample 8 unique polygons per bucket to reach ~144 targets (close to our 135 target)
print("Sampling 8 polygons from each populated bucket...")

sampling_targets = modern_events.groupby(['area_bin', 'urban_bin', 'mechanism_bin'], observed=False).apply(
    lambda x: x.sample(n=min(len(x), 8), random_state=42)
).reset_index(drop=True)

# --- 5. SAVE FOR GEE ---
# We keep only the IDs, Geometry (WKT), and the bins for the GEE phase
cols_for_gee = [
    'event_id', 'geometry_wkt', 'area_km2', 'urban_percentage', 
    'PFDI_p99', 'area_bin', 'urban_bin', 'mechanism_bin'
]

final_targets = sampling_targets[cols_for_gee]
final_targets.to_pickle(TARGETS_PATH)

print(f"\nSUCCESS!")
print(f"Total polygons selected: {len(final_targets)}")
print(f"Targets saved to: {TARGETS_PATH}")

# Quick Check: Pivot table to see the coverage

print(final_targets.groupby(['area_bin', 'urban_bin', 'mechanism_bin'], observed=False).size())
```

```python
import pandas as pd
import geopandas as gpd
from shapely import wkt
import folium
from branca.element import Template, MacroElement

# --- 1. LOAD DATA ---
df = pd.read_pickle(TARGETS_PATH)
df['geometry'] = df['geometry_wkt'].apply(wkt.loads)
gdf = gpd.GeoDataFrame(df, geometry='geometry', crs="EPSG:4326")

# --- 2. BASE MAP ---
map_center = [31.5, 34.8] # Centered on Israel
m = folium.Map(location=map_center, zoom_start=8, tiles='CartoDB positron')

# --- 3. COLORS & LAYERS ---
color_map = {'Low-Urban': 'green', 'Urban': 'orange', 'Highly-Urban': 'red'}

for _, row in gdf.iterrows():
    color = color_map.get(row['urban_bin'], 'blue')
    popup_text = f"<b>ID:</b> {row['event_id']}<br><b>Urban:</b> {row['urban_percentage']:.1f}%<br><b>Mechanism:</b> {row['mechanism_bin']}"
    
    folium.GeoJson(
        row['geometry'],
        style_function=lambda x, color=color: {
            'fillColor': color, 'color': color, 'weight': 2, 'fillOpacity': 0.6
        },
        popup=folium.Popup(popup_text, max_width=250)
    ).add_to(m)

# --- 4. ADDING THE LEGEND (HTML/CSS Injection) ---
legend_html = """
{% macro html(this, kwargs) %}
<div style="
    position: fixed; 
    bottom: 50px; left: 50px; width: 160px; height: 110px; 
    background-color: white; border:2px solid grey; z-index:9999; font-size:14px;
    padding: 10px;
    ">
    <b>Urban Density</b><br>
    <i style="background: green; width: 10px; height: 10px; float: left; margin-right: 5px; margin-top: 5px;"></i> Low (<20%)<br>
    <i style="background: orange; width: 10px; height: 10px; float: left; margin-right: 5px; margin-top: 5px;"></i> Urban (20-60%)<br>
    <i style="background: red; width: 10px; height: 10px; float: left; margin-right: 5px; margin-top: 5px;"></i> Highly Urban (>60%)
</div>
{% endmacro %}
"""

legend = MacroElement()
legend._template = Template(legend_html)
m.get_root().add_child(legend)

m
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# --- CONFIGURATION ---
CHRONICLE_DB_PATH = r"D:\Development\RESEARCH\urban_flood_database\chronicle\imerg_rain_outputs\chronicle_urban_df_with_IMERG_FULL.pkl"

# 1. Load Data
df = chronicle_df

# 2. Normalize and Clean
# Prepare the three versions in km2
df_plot = pd.DataFrame({
    'Original (Chronicle)': df['area_km2'],
    'Script 01 (GHS/Mollweide)': df['polygon_total_area_m2'] / 1e6,
    'Script 02 (GEE/Geodesic)': df['poly_area_km2']
})

# Remove zeros or NaNs to allow Log Scale plotting
df_plot = df_plot[(df_plot > 0).all(axis=1)].dropna()

# 3. Create Visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# --- Plot 1: PDF (Probability Density Function) ---
# Using KDE with log_scale=True to see the distribution shape clearly
sns.kdeplot(data=df_plot, ax=ax1, log_scale=True, common_norm=False, linewidth=2.5)
ax1.set_title("PDF: Area Distribution Comparison (Log Scale)", fontsize=14)
ax1.set_xlabel("Area (km)", fontsize=12)
ax1.set_ylabel("Density", fontsize=12)
ax1.grid(True, which="both", ls="-", alpha=0.2)

# --- Plot 2: CDF (Cumulative Distribution Function) ---
# Shows what percentage of the data is below a certain area size
sns.ecdfplot(data=df_plot, ax=ax2, log_scale=True, linewidth=2.5)
ax2.set_title("CDF: Cumulative Area Distribution (Log Scale)", fontsize=14)
ax2.set_xlabel("Area (km)", fontsize=12)
ax2.set_ylabel("Cumulative Probability", fontsize=12)
ax2.grid(True, which="both", ls="-", alpha=0.2)

plt.tight_layout()
plt.show()

# 4. Quick divergence check
print("\n--- Summary Statistics (km) ---")
print(df_plot.describe(percentiles=[0.1, 0.5, 0.9, 0.99]))


# 4. Quick divergence check (pretty print)
print("\n--- Summary Statistics (km) ---")

summary_stats = df_plot.describe(percentiles=[0.1, 0.5, 0.9, 0.99]).T

summary_stats = summary_stats.rename(columns={
    "count": "Count",
    "mean": "Mean",
    "std": "Std",
    "min": "Min",
    "10%": "P10",
    "50%": "Med",
    "90%": "P90",
    "99%": "P99",
    "max": "Max"
})

print(summary_stats.round(2).to_string())
```

```python
# Ensure datetime format
if not pd.api.types.is_datetime64_any_dtype(chronicle_df['start_time']):
    chronicle_df['start_time'] = pd.to_datetime(chronicle_df['start_time'], unit='s')

# Filter for the Modern Era
modern_events_df = chronicle_df[chronicle_df['start_time'] >= MODERN_ERA_START].copy()

# 2. Stratification Logic (Small Scale)
# Creating 3x3 buckets based on quantiles
modern_events_df['urban_level'] = pd.qcut(modern_events_df['urban_percentage'], 3, labels=['Low', 'Med', 'High'])
modern_events_df['upa_level'] = pd.qcut(modern_events_df['upa_p99'], 3, labels=['Low', 'Med', 'High'])

# FEASIBILITY SAMPLE: Pick only 5 polygons per bucket
test_targets_df = modern_events_df.groupby(['urban_level', 'upa_level']).apply(
    lambda x: x.sample(n=min(len(x), 5), random_state=42)
).reset_index(drop=True)

print(f"Feasibility sample created with {len(test_targets_df)} target polygons.")

# 3. Visualizing the Sample Distribution
plt.figure(figsize=(10, 6))
sns.scatterplot(data=test_targets_df, x='urban_percentage', y='upa_p99', hue='urban_level', style='upa_level', s=100)
plt.title("Feasibility Sample: Urban % vs UPA_P99")
plt.show()

# 4. Global Mapping (Small Scale)
test_targets_df['geometry'] = test_targets_df['geometry_wkt'].apply(shapely.wkt.loads)
gdf_test = gpd.GeoDataFrame(test_targets_df, geometry='geometry', crs="EPSG:4326")

world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
fig, ax = plt.subplots(figsize=(15, 7))
world.plot(ax=ax, color='lightgrey', edgecolor='white')
gdf_test.centroid.plot(ax=ax, color='red', markersize=50, label='Test Targets')
plt.title("Global Distribution of Feasibility Targets (N=45)")
plt.legend()
plt.show()

# 5. Inspect the data
print("\n--- Detailed View of the First 10 Test Targets ---")
display(test_targets_df[['event_id', 'urban_percentage', 'upa_p99', 'urban_level', 'upa_level']].head(45))
```
