# 04_chronicle_no_flood_stratified_sampling

## Step 1: Create 18 refined buckets (3x3x2) for targeted sampling.
## Action: Pick representative polygons for GEE No-Flood search.

### Imports

```python
# Standard library imports
import concurrent.futures
import json
import os
import uuid

# Third-party library imports
import ee
import numpy as np
import pandas as pd
import rasterio.features
import shapely.geometry
import shapely.wkt
from affine import Affine
```

```python


# --- 1. CONFIGURATION ---
# Path to the master dataset produced by script 03b
INPUT_MASTER_PATH = r"\\vscifs\hydrolab1\hydrolab\home\Raz\Research\chronicle\chronicle_rain_master_V07.pkl"
# Path to save the sampled targets for Google Earth Engine (GEE)
OUTPUT_DIR = r"D:\Development\RESEARCH\urban_flood_database\chronicle\no_flood_event_outputs"
TARGETS_PATH = os.path.join(OUTPUT_DIR, "sampling_targets_for_gee.pkl")

if not os.path.exists(OUTPUT_DIR):
    os.makedirs(OUTPUT_DIR)

# --- 2. LOAD DATA ---
print(f"Loading master dataset: {INPUT_MASTER_PATH}")
master_df = pd.read_pickle(INPUT_MASTER_PATH)

# Ensure start_time is treated as a datetime object with seconds unit for accurate filtering
master_df['start_time'] = pd.to_datetime(master_df['start_time'], unit='s', errors='coerce')

# Debugging: Print available date range to verify filtering logic
print(f"Date range in dataset: {master_df['start_time'].min()} to {master_df['start_time'].max()}")

# Filter for events from 2018 onwards to ensure high-quality satellite and urban data
modern_events = master_df[master_df['start_time'] >= pd.Timestamp('2018-01-01')].copy()

# Safety Check: If filter returns empty, fallback to full dataset and warn user
if len(modern_events) == 0:
    print("WARNING: No events found after 2018! Using full dataset instead.")
    modern_events = master_df.copy()

print(f"Events available for sampling: {len(modern_events)}")

# --- 3. REFINED STRATIFICATION (18 BUCKETS) ---
print("Categorizing into 18 strategic buckets...")

# A. Area Stratification (Tertiles)
# Using rank(method='first') to handle cases with identical area values and avoid qcut errors
modern_events['area_bin'] = pd.qcut(modern_events['area_km2'].rank(method='first'), 3, labels=['Small', 'Medium', 'Large'])

# B. Urban Percentage Stratification (Manual Thresholds)
# 0-20% (Low), 20-60% (Urban), 60-100% (Highly Urban)
urban_bins = [-0.01, 20, 60, 100.01]
urban_labels = ['Low-Urban', 'Urban', 'Highly-Urban']
modern_events['urban_bin'] = pd.cut(modern_events['urban_percentage'], bins=urban_bins, labels=urban_labels)

# C. Flood Mechanism Stratification (PFDI Threshold)
# < 1 (Pluvial-leaning), >= 1 (Fluvial/Extreme candidate)
pfdi_bins = [-0.01, 1, np.inf]
pfdi_labels = ['Pluvial (PFDI<1)', 'Fluvial (PFDI>=1)']
modern_events['mechanism_bin'] = pd.cut(modern_events['PFDI_p99'], bins=pfdi_bins, labels=pfdi_labels)

# --- 4. STRATIFIED SAMPLING ---
# Objective: Sample 8 unique polygons per bucket to reach approx 144 targets
print("Sampling up to 8 polygons from each populated bucket...")

# Remove rows with missing categorization values before grouping
modern_events = modern_events.dropna(subset=['area_bin', 'urban_bin', 'mechanism_bin'])

# Apply the sampling logic per bucket
sampling_targets = modern_events.groupby(['area_bin', 'urban_bin', 'mechanism_bin'], observed=False).apply(
    lambda x: x.sample(n=min(len(x), 8), random_state=42)
).reset_index(drop=True)

# --- 5. CLEANUP & SAVE FOR GEE ---
# Keep only essential columns for the GEE processing phase
cols_for_gee = [
    'event_id', 'geometry_wkt', 'area_km2', 'urban_percentage', 
    'PFDI_p99', 'area_bin', 'urban_bin', 'mechanism_bin'
]

# Ensure we only select columns that exist in the final dataframe
existing_cols = [c for c in cols_for_gee if c in sampling_targets.columns]
final_targets = sampling_targets[existing_cols]

# Save the final sampled list as a pickle file
final_targets.to_pickle(TARGETS_PATH)

print(f"\nSUCCESS! Total polygons selected: {len(final_targets)}")
print(f"Targets saved to: {TARGETS_PATH}")

# --- 6. SUMMARY TABLE ---
# Displaying the count of samples per category to verify distribution
print("\nFinal Sampling Distribution (Area & Urban vs Mechanism):")
print(final_targets.groupby(['area_bin', 'urban_bin', 'mechanism_bin'], observed=False).size().unstack(level=-1))
```

## Step 2: Create 18 refined buckets (3x3x2) for targeted sampling.
## 1. Automatically iterate over all targets in configurable batch sizes.
## 2. Skip already processed batches to allow easy resuming.
## 3. Extract all valid distinct storms per polygon using a 3-day buffer.
## 4. Implement strict TIMEOUTS to prevent GEE silent server hangs.

```python


# Initialize Google Earth Engine
# Run ee.Authenticate() first if you haven't authenticated this environment yet
try:
    ee.Initialize()
    print("Earth Engine initialized successfully.")
except Exception as e:
    print("Please authenticate Earth Engine first.")
    raise e

# --- 1. CONFIGURATION ---
TARGETS_PATH = r"D:\Development\RESEARCH\urban_flood_database\chronicle\no_flood_event_outputs\sampling_targets_for_gee.pkl"

# --- 2. LOAD ONE TARGET POLYGON ---
print("Loading sampling targets...")
targets_df = pd.read_pickle(TARGETS_PATH)

# Select the very first polygon for our symbolic test
test_row = targets_df.iloc[0]
event_id = test_row['event_id']
wkt_geom = test_row['geometry_wkt']

print(f"Testing extraction for Polygon ID: {event_id}")

# Convert WKT string to a format Earth Engine understands (GeoJSON coordinate list)
shapely_geom = shapely.wkt.loads(wkt_geom)

# Create an Earth Engine geometry object from the Shapely polygon bounds or exterior coordinates
# Note: For simplicity in the test, we use the bounding box. 
# In the full code, exact geometries can be passed.
bounds = shapely_geom.bounds
ee_polygon = ee.Geometry.Rectangle(bounds)

# --- 3. GEE RAINFALL EXTRACTION (SERVER-SIDE FILTERING) ---
START_DATE = '2020-01-01'
END_DATE = '2020-12-31'

print(f"Querying IMERG V07 for precipitation between {START_DATE} and {END_DATE}...")

# Load IMERG V07 collection
imerg = ee.ImageCollection("NASA/GPM_L3/IMERG_V07") \
    .select('precipitation') \
    .filterDate(START_DATE, END_DATE)

# Function to calculate maximum precipitation and attach it as a new property
def calculate_max_rain(image):
    max_dict = image.reduceRegion(
        reducer=ee.Reducer.max(),
        geometry=ee_polygon,
        scale=1000, 
        maxPixels=1e9
    )
    rain_val = max_dict.get('precipitation')
    # Return the image with the calculated value stored as a property
    return image.set('max_rain_mm_hr', rain_val)

# Apply the calculation to all 17,500+ images in the year
processed_imerg = imerg.map(calculate_max_rain)

# SERVER-SIDE FILTER: Keep only images where the max rain is greater than 2 mm/hr
# This drastically reduces the output from ~17,500 to just the significant rainy moments
rainy_events = processed_imerg.filter(ee.Filter.gt('max_rain_mm_hr', 2))

# Convert the filtered images to a simple feature collection for extraction
def create_feature(image):
    return ee.Feature(None, {
        'timestamp': image.get('system:time_start'),
        'max_rain_mm_hr': image.get('max_rain_mm_hr')
    })

final_features = rainy_events.map(create_feature)

print("Executing server-side calculations. This may take a moment...")
# Now getInfo() will only download the filtered records, safely under the 5000 limit
results = final_features.getInfo()

# --- 4. PROCESS AND DISPLAY RESULTS ---
rain_data = []

# Parse the safely filtered results
for feature in results.get('features', []):
    props = feature['properties']
    rain_val = props.get('max_rain_mm_hr')
    timestamp = props.get('timestamp')
    
    if rain_val is not None and timestamp is not None:
        rain_data.append({
            'datetime': pd.to_datetime(timestamp, unit='ms'),
            'max_rain_mm_hr': rain_val
        })

rain_df = pd.DataFrame(rain_data)

if not rain_df.empty:
    # Find the absolute peak rain event in this year
    peak_event = rain_df.loc[rain_df['max_rain_mm_hr'].idxmax()]
    print("\n--- TEST SUCCESSFUL ---")
    print(f"Found {len(rain_df)} significant rainfall intervals (>2 mm/hr).")
    print(f"Absolute peak rainfall detected:")
    print(f"Date & Time: {peak_event['datetime']}")
    print(f"Intensity: {peak_event['max_rain_mm_hr']:.2f} mm/hr")
else:
    print("\nNo significant rainfall (>2 mm/hr) found in this year/location.")
```

```python


# Initialize Google Earth Engine
try:
    ee.Initialize()
except Exception:
    ee.Authenticate()
    ee.Initialize()

# --- 1. CONFIGURATION ---
TARGETS_PATH = r"D:\Development\RESEARCH\urban_flood_database\chronicle\no_flood_event_outputs\sampling_targets_for_gee.pkl"
MASTER_PATH = r"\\vscifs\hydrolab1\hydrolab\home\Raz\Research\chronicle\chronicle_rain_master_V07.pkl"
OUTPUT_DIR = r"D:\Development\RESEARCH\urban_flood_database\chronicle\no_flood_event_outputs"

BATCH_SIZE = 10
SEARCH_YEARS = [2018, 2019, 2020, 2021, 2022, 2023]
THRESH_1H = 7.0   
THRESH_24H = 1.0  
BUFFER_DAYS = 3   
SCALE = 0.1

REQUIRED_COLUMNS = [
    'uuid', 'event_id', 'is_flood', 'area_km2', 'start_time', 'end_time', 'duration_days',
    'geometry_wkt', 'urban_built_up_area_m2', 'urban_percentage',
    'upa_max', 'upa_p95', 'upa_p99', 'PFDI_p95', 'PFDI_p99',
    'PFDI_max', 'imerg_matrix', 'imerg_mask', 'imerg_meta', 'imerg_type',
    '30_max_rainfall_intens', '60_max_rainfall_intens',
    '120_max_rainfall_intens', '240_max_rainfall_intens',
    '360_max_rainfall_intens', '720_max_rainfall_intens',
    '1440_max_rainfall_intens'
]

# --- 2. TIMEOUT HELPER FUNCTION ---
def get_info_with_timeout(ee_object, timeout_seconds):
    """Executes GEE getInfo() with a strict timeout using ThreadPoolExecutor."""
    with concurrent.futures.ThreadPoolExecutor(max_workers=1) as executor:
        future = executor.submit(ee_object.getInfo)
        try:
            return future.result(timeout=timeout_seconds)
        except concurrent.futures.TimeoutError:
            print(f"\n      [!] TIMEOUT: GEE server did not respond within {timeout_seconds}s. Skipping.")
            return None
        except Exception as e:
            print(f"\n      [!] ERROR during GEE request: {e}")
            return None

# --- 3. LOAD DATA ---
print("Loading master datasets...")
sampled_polygons_df = pd.read_pickle(TARGETS_PATH)
flood_events_database_df = pd.read_pickle(MASTER_PATH)
flood_events_database_df['start_time'] = pd.to_datetime(flood_events_database_df['start_time'], errors='coerce')
total_polygons = len(sampled_polygons_df)

def fetch_raw_pixels_getregion(year, ee_search_geometry):
    start_date = f'{year}-01-01'
    end_date = f'{year}-12-31'
    imerg_collection = ee.ImageCollection("NASA/GPM_L3/IMERG_V07").select('precipitation').filterDate(start_date, end_date)
    region_request = imerg_collection.getRegion(ee_search_geometry, 10000)
    # Apply a 120-second timeout for the year scanning
    return get_info_with_timeout(region_request, timeout_seconds=120)

# --- 4. AUTO-BATCHING LOOP ---
for start_idx in range(0, total_polygons, BATCH_SIZE):
    end_idx = min(start_idx + BATCH_SIZE, total_polygons)
    output_filename = os.path.join(OUTPUT_DIR, f"no_flood_batch_{start_idx}_to_{end_idx}.pkl")
    
    print(f"\n=====================================================================")
    print(f"STARTING BATCH: Polygons {start_idx} to {end_idx} (Total targets: {total_polygons})")
    print(f"=====================================================================")
    
    if os.path.exists(output_filename):
        print(f"Batch {start_idx}-{end_idx} already exists at {output_filename}. Skipping to next batch.")
        continue
        
    polygons_to_process = sampled_polygons_df.iloc[start_idx:end_idx]
    all_generated_no_flood_events = []
    
    for relative_idx, (_, polygon_row) in enumerate(polygons_to_process.iterrows()):
        current_event_id = polygon_row['event_id']
        print(f"\n--- Processing Polygon {relative_idx+1}/{len(polygons_to_process)} (ID: {current_event_id}) ---")
        
        original_flood_record = flood_events_database_df[flood_events_database_df['event_id'] == current_event_id].iloc[0]
        original_flood_date = original_flood_record['start_time']
        wkt_geometry_string = polygon_row['geometry_wkt']
        
        shapely_geometry = shapely.wkt.loads(wkt_geometry_string)
        polygon_bounds = shapely_geometry.bounds
        try:
            ee_polygon_exact = ee.Geometry.Polygon(list(shapely_geometry.exterior.coords))
        except Exception:
            ee_polygon_exact = ee.Geometry.BBox(*polygon_bounds)

        search_bounds = ee_polygon_exact.buffer(5000).bounds()
        valid_storm_peaks_for_polygon = []
        
        for current_year in SEARCH_YEARS:
            try:
                raw_region_data = fetch_raw_pixels_getregion(current_year, search_bounds)
                if not raw_region_data or len(raw_region_data) <= 1:
                    continue
                    
                df_raw_pixels = pd.DataFrame(raw_region_data[1:], columns=raw_region_data[0])
                df_raw_pixels = df_raw_pixels.dropna(subset=['precipitation'])
                if df_raw_pixels.empty:
                    continue
                    
                df_raw_pixels['datetime'] = pd.to_datetime(df_raw_pixels['time'], unit='ms')
                df_raw_pixels['precipitation'] = pd.to_numeric(df_raw_pixels['precipitation'])
                
                precipitation_timeseries_df = df_raw_pixels.groupby('datetime')['precipitation'].max().to_frame(name='rain')
                precipitation_timeseries_df = precipitation_timeseries_df.resample('30min').max().fillna(0.0)
                
                precipitation_timeseries_df['roll_1h'] = precipitation_timeseries_df['rain'].rolling(window=2).mean() 
                precipitation_timeseries_df['roll_24h'] = precipitation_timeseries_df['rain'].rolling(window=48).mean() 
                
                if pd.notna(original_flood_date):
                    buffer_start = original_flood_date - pd.Timedelta(days=BUFFER_DAYS)
                    buffer_end = original_flood_date + pd.Timedelta(days=BUFFER_DAYS)
                    mask = (precipitation_timeseries_df.index >= buffer_start) & (precipitation_timeseries_df.index <= buffer_end)
                    precipitation_timeseries_df.loc[mask, 'roll_1h'] = 0.0
                    precipitation_timeseries_df.loc[mask, 'roll_24h'] = 0.0

                valid_storm_candidates = precipitation_timeseries_df[
                    (precipitation_timeseries_df['roll_1h'] > THRESH_1H) & 
                    (precipitation_timeseries_df['roll_24h'] > THRESH_24H)
                ]
                
                while not valid_storm_candidates.empty:
                    current_peak = valid_storm_candidates['roll_1h'].idxmax()
                    valid_storm_peaks_for_polygon.append(current_peak)
                    
                    storm_buffer_start = current_peak - pd.Timedelta(days=BUFFER_DAYS)
                    storm_buffer_end = current_peak + pd.Timedelta(days=BUFFER_DAYS)
                    mask = (precipitation_timeseries_df.index >= storm_buffer_start) & (precipitation_timeseries_df.index <= storm_buffer_end)
                    precipitation_timeseries_df.loc[mask, 'roll_1h'] = 0.0
                    precipitation_timeseries_df.loc[mask, 'roll_24h'] = 0.0
                    
                    valid_storm_candidates = precipitation_timeseries_df[
                        (precipitation_timeseries_df['roll_1h'] > THRESH_1H) & 
                        (precipitation_timeseries_df['roll_24h'] > THRESH_24H)
                    ]
            except Exception:
                pass 
                
        if not valid_storm_peaks_for_polygon:
            print(f"No storms found for ID {current_event_id}.")
            continue
        
        print(f"Downloading {len(valid_storm_peaks_for_polygon)} 3D matrices...")
        for storm_peak_datetime in valid_storm_peaks_for_polygon:
            try:
                storm_start_time = storm_peak_datetime - pd.Timedelta(hours=72)
                storm_end_time = storm_peak_datetime + pd.Timedelta(hours=24)
                region_of_interest = ee.Geometry.BBox(*polygon_bounds)
                
                imerg_3d_collection = ee.ImageCollection("NASA/GPM_L3/IMERG_V07").select('precipitation').filterBounds(region_of_interest).filterDate(storm_start_time, storm_end_time)
                data_version = 'final'
                
                # Timeout on the size check (lightweight, but still hits the server)
                size_request = imerg_3d_collection.size()
                collection_size = get_info_with_timeout(size_request, timeout_seconds=60)
                
                if collection_size is None:
                    continue # Skip this storm if server hangs on size check
                    
                if collection_size == 0:
                    data_version = 'late'
                    imerg_3d_collection = ee.ImageCollection("NASA/GPM_L3/IMERG_V07_LATE").select('precipitation').filterBounds(region_of_interest).filterDate(storm_start_time, storm_end_time)
                    
                image_stack = imerg_3d_collection.toBands()
                
                # Apply a strict 180-second timeout for the heavy 3D matrix extraction
                rect_request = image_stack.sampleRectangle(region=region_of_interest, defaultValue=0)
                rect_info = get_info_with_timeout(rect_request, timeout_seconds=180)
                
                if rect_info is None:
                    continue # Skip this matrix if server hangs
                    
                raw_pixels_dict = rect_info['properties']
                
                band_keys_sorted = sorted(list(raw_pixels_dict.keys()))
                precipitation_3d_matrix = np.stack([np.array(raw_pixels_dict[b], dtype=np.float32) for b in band_keys_sorted])
                
                matrix_height, matrix_width = precipitation_3d_matrix.shape[1], precipitation_3d_matrix.shape[2]
                min_lon, min_lat, max_lon, max_lat = polygon_bounds
                spatial_transform = Affine(SCALE, 0, min_lon, 0, -SCALE, max_lat)
                
                matrix_metadata = {'origin_top_left': (max_lat, min_lon), 'scale': SCALE, 'shape': (matrix_height, matrix_width), 'timestamps': band_keys_sorted}
                exact_polygon_mask = rasterio.features.rasterize([(shapely_geometry, 1)], out_shape=(matrix_height, matrix_width), transform=spatial_transform, fill=0, all_touched=True, dtype=np.uint8)
                
                masked_precipitation_matrix = precipitation_3d_matrix * exact_polygon_mask[np.newaxis, :, :]
                total_mask_pixels = exact_polygon_mask.sum()
                
                if total_mask_pixels == 0:
                    continue
                    
                spatial_mean_timeseries = masked_precipitation_matrix.sum(axis=(1,2)) / total_mask_pixels
                event_precipitation_series = pd.Series(spatial_mean_timeseries)
                
                calculated_intensities = {
                    '30_max_rainfall_intens': event_precipitation_series.max(),
                    '60_max_rainfall_intens': event_precipitation_series.rolling(2).mean().max(),
                    '120_max_rainfall_intens': event_precipitation_series.rolling(4).mean().max(),
                    '240_max_rainfall_intens': event_precipitation_series.rolling(8).mean().max(),
                    '360_max_rainfall_intens': event_precipitation_series.rolling(12).mean().max(),
                    '720_max_rainfall_intens': event_precipitation_series.rolling(24).mean().max(),
                    '1440_max_rainfall_intens': event_precipitation_series.rolling(48).mean().max()
                }
                
                new_no_flood_record = original_flood_record.copy()
                new_no_flood_record['uuid'] = str(uuid.uuid4())
                new_no_flood_record['is_flood'] = 0  
                new_no_flood_record['start_time'] = storm_start_time
                new_no_flood_record['end_time'] = storm_end_time
                new_no_flood_record['duration_days'] = 4.0
                new_no_flood_record['imerg_matrix'] = precipitation_3d_matrix
                new_no_flood_record['imerg_mask'] = exact_polygon_mask
                new_no_flood_record['imerg_meta'] = matrix_metadata
                new_no_flood_record['imerg_type'] = f"{data_version}_no_flood"
                
                for key, value in calculated_intensities.items():
                    new_no_flood_record[key] = value
                    
                for col in REQUIRED_COLUMNS:
                    if col not in new_no_flood_record:
                        new_no_flood_record[col] = np.nan
                        
                all_generated_no_flood_events.append(new_no_flood_record[REQUIRED_COLUMNS])
            
            except Exception:
                pass
                
    # --- 5. SAVE BATCH ---
    if all_generated_no_flood_events:
        final_no_flood_df = pd.DataFrame(all_generated_no_flood_events)
        final_no_flood_df.to_pickle(output_filename)
        print(f"\n[v] Saved Batch {start_idx}-{end_idx} with {len(final_no_flood_df)} events.")
    else:
        print(f"\n[x] No events generated in Batch {start_idx}-{end_idx}.")

print("\nALL BATCHES COMPLETED SUCCESSFULLY.")
```
